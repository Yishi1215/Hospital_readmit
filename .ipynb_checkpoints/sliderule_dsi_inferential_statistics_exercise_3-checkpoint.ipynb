# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
from mpl_toolkits.axes_grid1 import make_axes_locatable
import scipy.stats as stats
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
```

****
## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

## Answer to question A: 

I don't agree with the analysis and recommendations mentioned in Preliminary Report. p-value is 3.7787539233400004e-23, which is less than the significance level of 0.01, we can reject the null hypothesis and conclude that mean excess readmission rate of hospitals/facilities with number of discharges < 100 is higher than that of hospitals/facilities with number of discharges > 1000. This is statistical significance. However, the practical significance is limited. The correlation between excess readmission rate and number of discharges is negative, but very weak. So the policy recommendations based on the statistical significance are not solid.

```python
clean_hospital_read_df = clean_hospital_read_df[['Number of Discharges','Excess Readmission Ratio']]
clean_hospital_read_df = clean_hospital_read_df.dropna(how='any')
```

```python
df_below100 = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] < 100]
df_above1000 = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] > 1000]
```

<br />In hospitals/facilities with number of discharges < 100, mean excess readmission rate $P_1$
<br />In hospitals/facilities with number of discharges > 1000, mean excess readmission rate $P_2$

** First, we will test the statement that rate of readmissions is trending down with increasing number of discharges.**
<br /> $H_0$：$P_1$ = $P_2$
<br /> $H_A$：$P_1$ < $P_2$
<br /> $\alpha = 0.01$  

```python
print(df_below100.shape)
print(df_above1000.shape)
```

```python
print('Mean excess readmission rate of hospitals/facilities with number of discharges < 100: ', 
      df_below100['Excess Readmission Ratio'].mean())
print('Mean excess readmission rate of hospitals/facilities with number of discharges > 1000: ',
      df_above1000['Excess Readmission Ratio'].mean())
```

```python
stats.f_oneway(df_below100['Excess Readmission Ratio'], df_above1000['Excess Readmission Ratio'])
```

```python
clean_hospital_read_df[['Number of Discharges','Excess Readmission Ratio']].corr()
```

```python
slope, intercept = np.polyfit(x, y, deg=1)
r = np.corrcoef(x, y)[0,1]
print('Slope:' , slope)
print('Intercept: ', intercept)
print('Pearson r: ', r)
```

## Answer to question B： 
The scatterplot that gave immediate visibility to the large cluster of data, it's easy to see the tendency and outliers. However, the plot might mislead the viewers with the red and green shaded area.

```python
y_pred = [i * slope + intercept for i in x]
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2, color='green')
plt.plot(x, y_pred, linewidth=3, color='red')
ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```
